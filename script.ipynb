```python
import json
```

```python
with open('server1.json') as server_1:
    data1 = json.load(server_1)
with open('server2.json') as server_2:
    data2 = json.load(server_2)
```

## Cherche doublons de user dans un fichier

### Fichier 1

```python
seen = set()
dups = {}
for x in data1:
    if x['user'] not in seen:
        seen.add(x['user'])
    else:
        if x['user'] in dups:
            dups[x['user']] = dups[x['user']] + 1
        else:
            dups[x['user']] = 2

sorted_dict = sorted(dups.items(), reverse=True, key=lambda x: x[1])
sorted_dict
```

### Fichier 2

```python
seen = set()
dups = {}
for x in data2:
    if x['user'] not in seen:
        seen.add(x['user'])
    else:
        if x['user'] in dups:
            dups[x['user']] = dups[x['user']] + 1
        else:
            dups[x['user']] = 2

sorted_dict = sorted(dups.items(), reverse=True, key=lambda x: x[1])
sorted_dict
```

## Users en commun sur les 2 fichiers

```python
seen = set()
communs = []
for x in data1:
    if x['user'] not in seen:
        seen.add(x['user'])

for x in data2:
    if x['user'] in seen:
        communs.append(x['user'])

print("nombre en communs:", len(communs), " sur :", len(seen))
communs
```
